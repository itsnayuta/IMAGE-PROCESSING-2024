**0. Import thư viện cần thiết**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

**1. Đọc và hiển thị ảnh gốc**

```python
image = cv2.imread('input_image.jpg', cv2.IMREAD_GRAYSCALE)  # Đọc ảnh ở chế độ mức xám
cv2_imshow(image)
```

**2. Xây dựng hàm biến đổi từ đó cân bằng ảnh**

```python
def histogram_equalization(image):
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    cdf = hist.cumsum()
    cdf_normalized = cdf * (255 / cdf[-1])
    cdf_normalized = np.uint8(cdf_normalized)
    equalized_image = cdf_normalized[image]
    return equalized_image
```

**3. Biến đổi ma trận mức xám ban đầu, từ đó cho ra ảnh mới**

```python
enhanced_image = histogram_equalization(image)
cv2.imwrite('enhanced_image.jpg', enhanced_image)
cv2_imshow(enhanced_image)
```
