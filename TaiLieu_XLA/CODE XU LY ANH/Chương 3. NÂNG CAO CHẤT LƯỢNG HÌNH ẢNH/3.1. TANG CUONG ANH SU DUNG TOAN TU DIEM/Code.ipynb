**0. Import thư viện cần thiết**

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

**1. Đọc và hiển thị ảnh gốc**

```python
image = cv2.imread('input_image.jpg', cv2.IMREAD_GRAYSCALE)  # Đọc ảnh ở chế độ mức xám
cv2_imshow(image)
```

**2. Xây dựng hàm biến đổi từ đó tăng cường ảnh**

```python
print(image.shape[0])
print(image.shape[1])
def enhance_pixel_value(r):
    if 0 <= r <= 3 * 255 / 8: s = (1/3) * r
    elif 3 * 255 / 8 < r <= 5 * 255 / 8: s = 3 * r - 255
    else:  s = (1/3) * r + 2 * 255 / 3
    if s < 0: s = 0
    if s > 255: s = 255
    return s
def enhance_image(image):
    enhanced_image = np.zeros_like(image)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]): enhanced_image[i, j] = enhance_pixel_value(image[i, j])
    return enhanced_image
```

**3. Biến đổi ma trận mức xám ban đầu, từ đó cho ra ảnh mới**

```python
enhanced_image = enhance_image(image)
cv2.imwrite('enhanced_image.jpg', enhanced_image)
cv2_imshow(enhanced_image)
```
